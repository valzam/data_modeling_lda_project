# Basic gensim topic modeling

## Create a 2d array with the words in each document

```python
long_texts = df[(df["text"].str.len() > 1000) & (df["language"] == "english")]
long_texts = long_texts.reset_index()
```

```python
RE_PUNCTUATION = '|'.join([re.escape(x) for x in string.punctuation])
documents_clean = long_texts["text"].str.replace(RE_PUNCTUATION,"")
documents_clean = documents_clean.str.replace('\n',"")
documents_clean = documents_clean.str.replace('‘',"")
documents_clean = documents_clean.str.replace('“',"")
documents_clean = documents_clean.str.replace('’',"")
documents_clean = documents_clean.str.replace('”',"")
documents_clean = documents_clean.str.replace('…',"")
documents_clean = documents_clean.str.replace('\x93',"")
documents_clean = documents_clean.str.replace('\x94',"")
documents_clean = documents_clean.str.replace('—',"")
documents_clean = documents_clean.str.replace('\x92',"")
stoplist = list(gensim.parsing.preprocessing.STOPWORDS) + ["<span"] + list(range(0,9))
def word_is_valid(word):
    return word not in stoplist and len(word) > 2 and len(word) < 20 and not "http" in word and not "www" in word \
                and not any(char.isdigit() for char in word)

texts = [[stemmer.stem(word) for word in document.lower().split() if word_is_valid(word)] for document in documents_clean]
```

## Generate a dictionary (word to id mapping)

```python
dictionary = gensim.corpora.Dictionary(texts)
print(dictionary)
```

```python
dictionary.filter_extremes(no_above=0.9)
print(dictionary)
```

```python
dictionary.save_as_text("dict_text.txt")
```

```python
dictionary[30232]
```

## Generate the Bag-of-words corpus

```python
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
len(corpus)
```

## LDA Model

```python
lda2 = gensim.models.LdaModel.load("lda_model.ge")
```

```python
lda = gensim.models.LdaModel(corpus[:-200], id2word=dictionary, num_topics=20, random_state=42)
top_words = [[word for word, _ in lda.show_topic(topicno, topn=50)] for topicno in range(lda.num_topics)]
for topics,i in zip(top_words, range(0,len(top_words))):
    print(str(i) + ": " + " ".join(topics[:10]))
```

```python
topics = []
perplexity = []

for i in tqdm(range(1,30)):
    lda = gensim.models.LdaModel(corpus[:-200], id2word=dictionary, num_topics=i, random_state=42)
    p = lda.log_perplexity(corpus[200:])
    topics.append(i)
    perplexity.append(p)

results = pd.Series(perplexity,index=topics)
```

```python
import pickle
pickle.dump(results, open("cv_perplexity_results.p", "wb"))
```

# HDP Model

```python
hdp = gensim.models.HdpModel(corpus, id2word=dictionary, random_state=42)
```

```python
len(hdp.show_topics())
```

```python
top_words = [[word for word, _ in hdp.show_topic(topicno, num_words=50)] for topicno in range(len(hdp.show_topics()))]
```

```python
for topics,i in zip(top_words, range(0,len(top_words))):
    print(str(i) + ": " + " ".join(topics[:10]))
```
